```
from fastai.data.all import *
from fastai.vision.all import *
```

```
path = Path('/content/drive/MyDrive/Dataset')
```

```

dls = ImageDataLoaders.from_folder(path, valid_pct=0.2, item_tfms=Resize(224),bs=8)
```

```
dls.valid_ds.items[:3]
```

```
dls.show_batch()
```

```
learn = cnn_learner(dls, models.resnet34, metrics=error_rate) # Initializing CNN
```

```
learn.model # RESNET-34 CNN architecture
```

```
learn.fit_one_cycle(40) # 4 epochs
```

```
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fit_one_cycle(20, 3e-3)
learn.unfreeze()
learn.fit_one_cycle(10, lr_max=slice(1e-6,1e-4))
```

```
learn.recorder.plot_loss()
```

```
interp = ClassificationInterpretation.from_learner(learn) # Initializing interpreter

losses,idxs = interp.top_losses() # Getting losses and class id's

```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```
learn.model_dir='/content/'
learn.save('stage-1') # saving model
learn.export('stage-1.pkl') # exports the model
```
